Perzeptrons erster schicht

```python
# Gewichte und bias-Werte sind in der Abbildung ablesbar und werden hier f√ºr die Perzeptrons der ersten Schicht dargestellt

def h1 (x_1):
  p1_w = [1, 0]
  p1_b = -2
  return np.sign(x[0]*p1_w[0] + x[1]*p1_w[1] + p1_b)

def h2 (x_2):
  p2_w = [0, 1]
  p2_b = -3
  return np.sign(x[0]*p2_w[0] + x[1]*p2_w[1] + p2_b)

print(h1(x))
print(h2(x))
```

Perzeptron zweiter Schicht

```python
# Das perzeptron der zweiten Schicht, ist eine AND/OR Abfrage der ersten schicht

p3_w = [1, 1]
p3_b = 1
```

```python
a = np.sign(h1(x)*p3_w[0] + h2(x)*p3_w[1] + p3_b)

print(a)
```
